```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
def print_circuit_values(circuit):
    simulator = Aer.get_backend('qasm_simulator')
    result = execute(circuit, backend=simulator, shots=1).result()
    counts = result.get_counts()
    print(counts)
```

```python
circuit = QuantumCircuit(2, 2, name="Switching from 1 to 0")

circuit.x(0)
circuit.z(1)
print_circuit_values(circuit)

circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.barrier()
print_circuit_values(circuit)

circuit.cx(0, 1)
print_circuit_values(circuit)

circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.barrier()
print_circuit_values(circuit)

circuit.h(0) # here the change is not measured, but the state changes - random change?
print_circuit_values(circuit)

circuit.barrier()
circuit.measure([0,1], [0,1]) # here the change is measured, so the states change - random change?
print_circuit_values(circuit)

circuit.draw(output='mpl')
```
