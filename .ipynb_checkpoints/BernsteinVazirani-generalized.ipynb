```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secretnumber = '1010'
```

```python
number_of_digits = len(secretnumber)
array_digit_position = range(number_of_digits)

circuit = QuantumCircuit(number_of_digits + 1, number_of_digits)

circuit.h(array_digit_position)
circuit.x(number_of_digits)
circuit.h(number_of_digits)

circuit.barrier()

for index, value in enumerate(reversed(secretnumber)):
    if (value == '1'):
        circuit.cx(index, number_of_digits)

circuit.barrier()

circuit.h(array_digit_position)

circuit.barrier()
circuit.measure(array_digit_position, array_digit_position)
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1).result()
counts = result.get_counts()
print(counts)
```
